# NLTK 

```python
#now we will convert the list into string to apply nltk
International_df['tags'] = International_df['tags'].apply(lambda x:" ".join(x))#converted to list
```

```python
import nltk
from nltk.stem.porter import PorterStemmer 
ps=PorterStemmer()
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
International_df['tags']=International_df['tags'].apply(stem)
```

## Cosine Similarity

##### The cosine similarity measures the similarity between vector lists by calculating the cosine angle between the two vector lists. If you consider the cosine function, its value at 0 degrees is 1 and -1 at 180 degrees.

### Cosine Similarity = (A.B) / (||A||.||B||) 

##### In case of cosine similarity the higher the resultant value higher the similarity but in case of cosine distance lower the resultant value higher the similarity as cosine distance =1-[cosine similarity]

```python
from sklearn.metrics.pairwise import cosine_similarity 
#means it will range from 0 to 1 as 1 means totally similar and 0 means
```

```python
similarity=cosine_similarity(vectors)
#similarity score=1 means exactly same and 0 means totally different in between propotional
```

# Main function

```python
#we will fetch top 5 similar movie

def recommend(movie):
    movie_index = International_df[International_df['title']== movie].index[0]
    distances=similarity[movie_index]
    movies_list= sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:6] 
    
    for i in movies_list:
        print(International_df.iloc[i[0]].title) 
        
    
    return
```
